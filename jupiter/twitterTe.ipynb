```python
def getOtherHashTag(texto):
  return ', '.join([ele for ele in texto.split() if '#' in ele])

def getContry(contry):
  contries = contry.split(',')
  if(len(contries) > 0):
    return contries[-1]
  return contries
```

```python
# palabras_validas = ['aislamiento', 'pandemia', 'virus', 'coronavirus']
query = {'q': 'coronavirus',
        'count': 999,
        'lang': 'es',
         #'result_type': 'popular', #'mixed'
         #'until': '2021-06-02'
        }
statuses = python_tweets.search(**query, include_rts=False,tweet_mode='extended')['statuses']
general = python_tweets.search(**query)
#general

```

```python
data = [];
for twit in statuses:
    data.append({
        'id': twit['id'],
        'usuario': twit['user']['screen_name'],
        'texto': twit['full_text'],
        'fecha': twit['created_at'],
        'ubicacion': twit['user']['location'],
        #'pais': getContry(twit['user']['location']),
        'cant_seguidos': twit['user']['friends_count'],
        'cant_seguidores': twit['user']['followers_count'],
        'cant_retweet': twit['retweet_count'],
        'cant_favoritos': twit['favorite_count'],
        'hash_relacionados': getOtherHashTag(twit['full_text']),
        
    })


dataFrame = pd.DataFrame(data)
ordered_by = ['cant_seguidos','cant_seguidos', 'cant_retweet']
accendi_by = [ False, False,False]
dataFrameOrde = dataFrame.sort_values(by=ordered_by, ascending=accendi_by)
```
